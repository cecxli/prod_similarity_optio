```python
import numpy as np
from sentence_transformers import SentenceTransformer
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA
```

```python
import plotly_express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
xls = pd.ExcelFile("./Data/ML Test Assignment - Similar Products Data.xlsx")
cat_df = pd.read_excel(xls,'Categories')
prod_df = pd.read_excel(xls,'Products')
prodsp_df = pd.read_excel(xls,'ProductSpecifications')
```

```python
prodsp_df.drop_duplicates(inplace=True)
```

```python
# prodsp_df.pivot(index='ProductId',columns = 'AttributeId')['AttributeValue']
```

```python
prod_df
```

```python
prod_df['Description'] = prod_df['Description'].map(str)
```

```python
text_data = prod_df['Description']
model = SentenceTransformer('distilbert-base-nli-mean-tokens')
embeddings = model.encode(text_data, show_progress_bar=True)
```

```python
embeddings.shape
```

```python
X = np.array(embeddings)
```

```python
pca = PCA()
t = pca.fit_transform(embeddings)
explained_variance = pca.explained_variance_ratio_.cumsum()
```

```python
fig = px.line(x = range(768), y = explained_variance,width=800,height=400)

fig.show()
```
